## **Define Model:**

```python
from langchain_groq import ChatGroq

# Initialize the Groq model
groq = ChatGroq(
    model="llama-3.1-70b-versatile",
    temperature=0,
)
```

# **Example On How To Use/Invoke Grok:**

```python
from langchain_core.messages import HumanMessage

msg: HumanMessage = HumanMessage(content="Hello there")
messages:list[HumanMessage] = [msg]
print(messages)
```

```python
try:
    response_groq = groq.invoke(messages)
    print(response_groq.content)
except Exception as e:
    print("An error occurred:", e)
```
